```python
from sklearn.model_selection import train_test_split

def split_data(df, target_col, test_size=0.2, random_state=42):
    X = df.drop(columns=[target_col])
    y = df[target_col]
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )
    return X_train, X_test, y_train, y_test
```

```python
df.info()
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report, roc_auc_score

# 1. Chuyá»ƒn Churn thÃ nh 0/1
df['Churn'] = df['Churn'].astype(int)

# 2. Chuyá»ƒn cá»™t State thÃ nh string (náº¿u chÆ°a)
df['State'] = df['State'].astype(str)

# 3. Gá»i láº¡i hÃ m báº¡n Ä‘Ã£ viáº¿t
X_train, X_test, y_train, y_test = split_data(df, target_col='Churn')

# 4. Khai bÃ¡o cá»™t categorical
cat_features = ['State']

# 5. Train CatBoost
model = CatBoostClassifier(
    iterations=200,
    learning_rate=0.1,
    depth=4,
    class_weights=[1, 5],  # Náº¿u lá»›p churn bá»‹ lá»‡ch
    verbose=0
)

model.fit(X_train, y_train, cat_features=cat_features)

# 6. ÄÃ¡nh giÃ¡
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# 7. XÃ¡c suáº¥t (Ä‘á»ƒ tÃ­nh ROC AUC)
y_train_prob = model.predict_proba(X_train)[:, 1]
y_test_prob = model.predict_proba(X_test)[:, 1]

# 8. ÄÃ¡nh giÃ¡ káº¿t quáº£
print("==== ðŸ“Š Train Set Evaluation ====")
print(classification_report(y_train, y_train_pred))
print(f"ROC AUC (Train): {roc_auc_score(y_train, y_train_prob):.4f}")

print("\n==== ðŸ“Š Test Set Evaluation ====")
print(classification_report(y_test, y_test_pred))
print(f"ROC AUC (Test): {roc_auc_score(y_test, y_test_prob):.4f}")

```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.model_selection import train_test_split

# HÃ m chia táº­p
def split_data(df, target_col, test_size=0.2, random_state=42):
    X = df.drop(columns=[target_col])
    y = df[target_col]
    return train_test_split(X, y, test_size=test_size, random_state=random_state)

# 1. Chuáº©n bá»‹ dá»¯ liá»‡u
df['Churn'] = df['Churn'].astype(int)
df['State'] = df['State'].astype(str)

# 2. Chia dá»¯ liá»‡u
X_train, X_test, y_train, y_test = split_data(df, target_col='Churn')

# 3. Cá»™t categorical
cat_features = ['State']

# 4. Train CatBoost
model = CatBoostClassifier(
    iterations=200,
    learning_rate=0.1,
    depth=4,
    class_weights=[1, 5],
    verbose=0
)
model.fit(X_train, y_train, cat_features=cat_features)

# 5. Dá»± Ä‘oÃ¡n xÃ¡c suáº¥t
y_test_prob = model.predict_proba(X_test)[:, 1]

# 6. Dá»± Ä‘oÃ¡n theo threshold = 0.5 (máº·c Ä‘á»‹nh)
y_pred_05 = (y_test_prob > 0.5).astype(int)
print("=== ðŸ“Š TEST SET (Threshold = 0.5 - Default) ===")
print(classification_report(y_test, y_pred_05))
print(f"ROC AUC: {roc_auc_score(y_test, y_test_prob):.4f}")

# 7. Dá»± Ä‘oÃ¡n theo threshold = 0.4
y_pred_04 = (y_test_prob > 0.4).astype(int)
print("\n=== ðŸ“Š TEST SET (Threshold = 0.4 - Boost Recall) ===")
print(classification_report(y_test, y_pred_04))
print(f"ROC AUC: {roc_auc_score(y_test, y_test_prob):.4f}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, recall_score
# Táº¡o CatBoost cÆ¡ báº£n (chÆ°a fit)
model = CatBoostClassifier(
    verbose=0,
    class_weights=[1, 5],  # Náº¿u dá»¯ liá»‡u lá»‡ch
    cat_features=['State']
)

# Grid cÃ¡c tham sá»‘ cáº§n thá»­
param_grid = {
    'depth': [4, 6, 8],
    'learning_rate': [0.01, 0.05, 0.1],
    'iterations': [100, 150]
}

# Scorer: láº¥y recall cá»§a lá»›p churn (label 1)
recall_scorer = make_scorer(recall_score, pos_label=1)

# Khá»Ÿi táº¡o GridSearchCV
grid = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring=recall_scorer,
    cv=3,
    n_jobs=-1  # DÃ¹ng Ä‘a luá»“ng cho nhanh
)

# Fit trÃªn táº­p huáº¥n luyá»‡n
grid.fit(X_train, y_train)

# In káº¿t quáº£ tá»‘t nháº¥t
print("ðŸŽ¯ MÃ´ hÃ¬nh tá»‘t nháº¥t:")
print(grid.best_params_)
print(f"Recall (CV trung bÃ¬nh): {grid.best_score_:.4f}")

# Láº¥y model tá»‘t nháº¥t
best_model = grid.best_estimator_
```

```python
# 1. GÃ¡n láº¡i best hyperparameters
best_model = CatBoostClassifier(
    depth=6,
    iterations=150,
    learning_rate=0.05,
    class_weights=[1, 5],
    cat_features=['State'],
    verbose=0
)

# 2. Huáº¥n luyá»‡n
best_model.fit(X_train, y_train)

# 3. Dá»± Ä‘oÃ¡n xÃ¡c suáº¥t
y_train_prob = best_model.predict_proba(X_train)[:, 1]
y_test_prob = best_model.predict_proba(X_test)[:, 1]

# 4. Dá»± Ä‘oÃ¡n nhÃ£n theo threshold máº·c Ä‘á»‹nh = 0.5
y_train_pred = (y_train_prob > 0.5).astype(int)
y_test_pred = (y_test_prob > 0.5).astype(int)

# 5. In káº¿t quáº£
print("=== ðŸ“Š TRAIN SET EVALUATION ===")
print(classification_report(y_train, y_train_pred))
print(f"ROC AUC (Train): {roc_auc_score(y_train, y_train_prob):.4f}")

print("\n=== ðŸ“Š TEST SET EVALUATION ===")
print(classification_report(y_test, y_test_pred))
print(f"ROC AUC (Test): {roc_auc_score(y_test, y_test_prob):.4f}")
```

```python
from sklearn.model_selection import train_test_split

def split_data(df, target_col, test_size=0.2, random_state=42):
    X = df.drop(columns=[target_col])
    y = df[target_col]
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )
    return X_train, X_test, y_train, y_test
```

```python
df.info()
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report, roc_auc_score

# 1. Chuyển Churn thành 0/1
df['Churn'] = df['Churn'].astype(int)

# 2. Chuyển cột State thành string (nếu chưa)
df['State'] = df['State'].astype(str)

# 3. Gọi lại hàm bạn đã viết
X_train, X_test, y_train, y_test = split_data(df, target_col='Churn')

# 4. Khai báo cột categorical
cat_features = ['State']

# 5. Train CatBoost
model = CatBoostClassifier(
    iterations=200,
    learning_rate=0.1,
    depth=4,
    class_weights=[1, 5],  # Nếu lớp churn bị lệch
    verbose=0
)

model.fit(X_train, y_train, cat_features=cat_features)

# 6. Đánh giá
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# 7. Xác suất (để tính ROC AUC)
y_train_prob = model.predict_proba(X_train)[:, 1]
y_test_prob = model.predict_proba(X_test)[:, 1]

# 8. Đánh giá kết quả
print("==== 📊 Train Set Evaluation ====")
print(classification_report(y_train, y_train_pred))
print(f"ROC AUC (Train): {roc_auc_score(y_train, y_train_prob):.4f}")

print("\n==== 📊 Test Set Evaluation ====")
print(classification_report(y_test, y_test_pred))
print(f"ROC AUC (Test): {roc_auc_score(y_test, y_test_prob):.4f}")

```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.model_selection import train_test_split

# Hàm chia tập
def split_data(df, target_col, test_size=0.2, random_state=42):
    X = df.drop(columns=[target_col])
    y = df[target_col]
    return train_test_split(X, y, test_size=test_size, random_state=random_state)

# 1. Chuẩn bị dữ liệu
df['Churn'] = df['Churn'].astype(int)
df['State'] = df['State'].astype(str)

# 2. Chia dữ liệu
X_train, X_test, y_train, y_test = split_data(df, target_col='Churn')

# 3. Cột categorical
cat_features = ['State']

# 4. Train CatBoost
model = CatBoostClassifier(
    iterations=200,
    learning_rate=0.1,
    depth=4,
    class_weights=[1, 5],
    verbose=0
)
model.fit(X_train, y_train, cat_features=cat_features)

# 5. Dự đoán xác suất
y_test_prob = model.predict_proba(X_test)[:, 1]

# 6. Dự đoán theo threshold = 0.5 (mặc định)
y_pred_05 = (y_test_prob > 0.5).astype(int)
print("=== 📊 TEST SET (Threshold = 0.5 - Default) ===")
print(classification_report(y_test, y_pred_05))
print(f"ROC AUC: {roc_auc_score(y_test, y_test_prob):.4f}")

# 7. Dự đoán theo threshold = 0.4
y_pred_04 = (y_test_prob > 0.4).astype(int)
print("\n=== 📊 TEST SET (Threshold = 0.4 - Boost Recall) ===")
print(classification_report(y_test, y_pred_04))
print(f"ROC AUC: {roc_auc_score(y_test, y_test_prob):.4f}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, recall_score
# Tạo CatBoost cơ bản (chưa fit)
model = CatBoostClassifier(
    verbose=0,
    class_weights=[1, 5],  # Nếu dữ liệu lệch
    cat_features=['State']
)

# Grid các tham số cần thử
param_grid = {
    'depth': [4, 6, 8],
    'learning_rate': [0.01, 0.05, 0.1],
    'iterations': [100, 150]
}

# Scorer: lấy recall của lớp churn (label 1)
recall_scorer = make_scorer(recall_score, pos_label=1)

# Khởi tạo GridSearchCV
grid = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring=recall_scorer,
    cv=3,
    n_jobs=-1  # Dùng đa luồng cho nhanh
)

# Fit trên tập huấn luyện
grid.fit(X_train, y_train)

# In kết quả tốt nhất
print("🎯 Mô hình tốt nhất:")
print(grid.best_params_)
print(f"Recall (CV trung bình): {grid.best_score_:.4f}")

# Lấy model tốt nhất
best_model = grid.best_estimator_
```

```python
# 1. Gán lại best hyperparameters
best_model = CatBoostClassifier(
    depth=6,
    iterations=150,
    learning_rate=0.05,
    class_weights=[1, 5],
    cat_features=['State'],
    verbose=0
)

# 2. Huấn luyện
best_model.fit(X_train, y_train)

# 3. Dự đoán xác suất
y_train_prob = best_model.predict_proba(X_train)[:, 1]
y_test_prob = best_model.predict_proba(X_test)[:, 1]

# 4. Dự đoán nhãn theo threshold mặc định = 0.5
y_train_pred = (y_train_prob > 0.5).astype(int)
y_test_pred = (y_test_prob > 0.5).astype(int)

# 5. In kết quả
print("=== 📊 TRAIN SET EVALUATION ===")
print(classification_report(y_train, y_train_pred))
print(f"ROC AUC (Train): {roc_auc_score(y_train, y_train_prob):.4f}")

print("\n=== 📊 TEST SET EVALUATION ===")
print(classification_report(y_test, y_test_pred))
print(f"ROC AUC (Test): {roc_auc_score(y_test, y_test_prob):.4f}")
```
